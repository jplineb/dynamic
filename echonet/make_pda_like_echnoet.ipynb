## Quick prototyping

```python
def extract_video_metadata(file_path):
    cv2video = cv2.VideoCapture(file_path)
    height = cv2video.get(cv2.CAP_PROP_FRAME_HEIGHT)
    width  = cv2video.get(cv2.CAP_PROP_FRAME_WIDTH)
    
    framecount = cv2video.get(cv2.CAP_PROP_FRAME_COUNT ) 
    fps = cv2video.get(cv2.CAP_PROP_FPS)
    
    return height, width, framecount, fps
```

## Modify function use for dataframe

```python
def extract_video_metadata_df(row):
    file_path = row["FileName"]
    cv2video = cv2.VideoCapture(file_path)
    print(f"Starting video {file_path}")
    row["FrameHeight"] = cv2video.get(cv2.CAP_PROP_FRAME_HEIGHT)
    row["FrameWidth"]  = cv2video.get(cv2.CAP_PROP_FRAME_WIDTH)
    
    row["NumberOfFrames"] = cv2video.get(cv2.CAP_PROP_FRAME_COUNT)
    row["FPS"] = cv2video.get(cv2.CAP_PROP_FPS)
    
    return row
```

```python
df_new_meta = new_df.apply(extract_video_metadata_df, axis=1)
```
